#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('SINASC_RO_2019.csv')
data.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
data.loc[data['APGAR5'] <= 3 , 'CATEGORIA'] = 'asfixia grave'
data.loc[(data['APGAR5'] > 3) & (data['APGAR5'] <=5) , 'CATEGORIA'] = 'asfixia moderada'
data.loc[(data['APGAR5'] > 5) & (data['APGAR5'] <=7), 'CATEGORIA'] = 'asfixia leve'
data.loc[data['APGAR5'] >=8 , 'CATEGORIA'] = 'normal'
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.catplot(
    x = 'GESTACAO'
    ,y = 'APGAR5'
    ,hue = 'SEXO'
    ,kind='bar'
    ,data = data
    ,errorbar = None
);
plt.xticks(rotation=90)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabcruz = pd.crosstab(data['GESTACAO'], data['APGAR5'])
tabcruz
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tabcruz_prctline = tabcruz.div(tabcruz.sum(axis=1), axis=0)
tabcruz_prctline
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
tabcruz_prctline.plot.bar();

# O conjunto de dados corrobora com a afirmação dos médicos, pois na gestação que tem 37 semanas
# ou mais a incidência de asfixia é baixíssima, já nas demais semanas o índice de asfixia varia muito,
# no caso aparece mais vezes os valores de APGAR5 de 0 a 7. Podendo ocasionar assim, complicações
# para a criança.
```
